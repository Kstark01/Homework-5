```r
#question 2a
id <- c(1, 2, 3, 4, 5)
name <- c("Peter", "Amy", "Ryan", "Gary", "Michelle")
salary <- c(623.30, 515.20, 611.00, 729.00, 843.25)

df <- data.frame(id, name, salary)

df

#question 2b
department <- c("Finance", "IT", "Human Resources", "Upper Management", "Sales")

df2 <- cbind(df, department)

df2

#question 2c
df[c(1,3,5), c(2, 3)]

```

```r
#question 2d
salary <- c(df$salary[c(1,4,5)])
name <- df$name[c(1,4,5)]


barplot(salary, xlab= "Name", ylab = "Salary", names.arg = name)

```

```r
#question 2e
values <- c(median(df$salary), max(df$salary), min(df$salary))
labs <- c("Median Salary", "Max Salary", "Min Salary")

values

pie(values, labels = labs, main = "Office Salaries")
```
